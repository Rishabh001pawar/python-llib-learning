```python
#repaet an array
arr=np.array([[1,2,3]])
r1=np.repeat(arr,3, axis=0)
print(r1)
```

```python
output=np.ones((5,5))
print(output)

z=np.zeros((3,3))
z[1,1]=9
print(z)

output[1:-1,1:-1]=z
print(output)
```

## Be careful when cpying arrays!!

```python
a=np.array([1,2,3])
b=a;
b[0]=100
print(b)
```

```python
a=np.array([1,2,3])
b=a.copy()
b[0]=100
print(b)
```

### Reorganinsing arrays

```python
before=np.array([[1,2,3,4],[5,6,7,8]])
print(before)

after=before.reshape((2,2))
print(after)
```

```python
# Vertically staking vectors

v1=np.array([1,2,3,4])
v2=np.array([5,6,7,8])

np.vstack([v1,v2,v1,v2])
```

```python
# Horizontal Stack

h1=np.ones((2,4))
h2=np.zeros((2,2))

np.hstack((h1,h2))
```
